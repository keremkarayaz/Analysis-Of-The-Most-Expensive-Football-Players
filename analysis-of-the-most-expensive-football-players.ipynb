```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df=df.drop(["Name","Position","Club"],axis=1,inplace=False)
X=df.drop(columns=["Country"])
Y=df["Country"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)

model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
predictions=model.predict(X_test)

score=accuracy_score(Y_test,predictions)
score
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Age"][:50] ,df["G\A"][:50])
plt.xlabel("Ages")
plt.ylabel("G\A")
plt.show()
```

```python
from sklearn.cluster import KMeans
df=df.drop(["Country"],axis=1,inplace=False)
t=df.copy()
kmeans=KMeans(2)
kmeans.fit(t)
```

```python
clusters=t.copy()
clusters["cluster_pred"]=kmeans.fit_predict(t)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(clusters["Age"],clusters["G\A"],c=clusters["cluster_pred"],cmap="rainbow")
plt.xlabel("Ages")
plt.ylabel("G\A")
plt.grid()
plt.show()
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(t)
x_scaled
```

```python
wcss=[]

for i in range(1,50):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,50),wcss)
plt.xlabel("K-Means")
plt.ylabel("WCSS")
plt.grid()
plt.show()
```

```python
kmeans_new=KMeans(8)
kmeans.fit(x_scaled)
cluster_new=t.copy()
cluster_new["cluster_pred"]=kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.figure(figsize=(10,6))
plt.scatter(cluster_new["Age"],cluster_new["G\A"],c=cluster_new["cluster_pred"],cmap="rainbow")
plt.xlabel("Ages")
plt.ylabel("G\A")
plt.grid()
plt.show()
```
